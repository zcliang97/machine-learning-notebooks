```python
# --- Gradient Descent Example ---

# find the min of the function y = (x + 5)^2
# expect to have -5

# --- Gradient Descent Algo ---
# iter1:
# X1 = X0 - learning_rate * (dy/dx)

# iter2:
# X2 = X1 - learning_Rate * (dy/dx)

# ...
```

```python
import matplotlib.pyplot as plt

# starting x
curr_x = 3

# learning rate - can vary to see distribution changes
learning_rate = 0.05

# precision before stop algo
precision = 0.000001
delta_step_size = 1

# iteration limits
max_iters = 10000
num_iters = 0

# eequations for the function
f = lambda x: (x + 5)**2
df = lambda x: 2*(x + 5)

# plot array
x, y = [], []


while num_iters < max_iters and delta_step_size > precision:
    prev_x = curr_x
    curr_x = prev_x - learning_rate * df(prev_x)
    delta_step_size = abs(curr_x - prev_x)
    num_iters += 1
    x.append(curr_x)
    y.append(f(curr_x))
    print(curr_x)

plt.plot(x, y, 'ro')
plt.plot(curr_x, f(curr_x), 'bx')
plt.show()

print("Final min_x value: ", curr_x)
```
